## Explore Rating Score Performance

```python
def rating_perf (x):
    if x <= 1:
        return 'Low'
    elif x <= 2:
        return 'Low-Medium'
    elif x <= 3:
        return 'Medium'
    elif x <= 4:
        return 'Medium-High'
    elif x >3:
        return 'High'
```

```python
dataset['rating_perf']=dataset['rating'].apply(rating_perf)
```

```python
rating_performance_category = dataset[['category_1','rating_perf','product_id']]
```

```python
rating_performance_category_pivot = pd.pivot_table(data=rating_performance_category
                                                  ,index='category_1'
                                                    ,columns=['rating_perf']
                                                  ,aggfunc='count'
                                                  ,fill_value=0)
```

```python
rating_performance_category_pivot = rating_performance_category_pivot.droplevel(0,axis=1)
```

```python
rating_performance_category_pivot = rating_performance_category_pivot[['High','Medium-High','Medium','Low-Medium']]
```

```python
rating_performance_heatmap = sns.heatmap(rating_performance_category_pivot,cmap='Greens')
plt.title('Rating Performance of Category 1')
plt.xlabel('Rating Performance')
plt.ylabel('Category 1')
```

## Explore Actual Price

```python
Pricing = pd.DataFrame(dataset['actual_price'])
```

```python
Pricing_Distribution_Hist = sns.histplot(data=Pricing,x='actual_price',bins=10,color='green')
Pricing_Distribution_Hist.bar_label(Pricing_Distribution_Hist.containers[0])
```

```python
def price_group(x):
    max_price = 200000
    range_size = 5000
    lower = 0
    count = 0
    for i in range(int(max_price/range_size)):
        upper = (5000*count) + range_size
        if x <= upper:
            return str(lower)+'-'+str(upper)
        lower = upper + 1
        count = count + 1
```

```python
Pricing['price_group'] = Pricing['actual_price'].apply(price_group)
```

```python
Price_Group_Distribution_countplot = sns.countplot(data=Pricing,x='price_group',color='green',order=['0-5000','5001-10000','10001-15000','15001-20000','20001-25000','25001-30000','30001-35000','35001-40000','40001-45000','45001-50000','50001-55000','55001-60000','60001-65000','65001-70000','70001-75000','75001-80000','80001-85000','85001-90000','90001-95000','95001-100000','100001-105000','105001-110000','110001-115000','115001-120000','120001-125000','125001-130000','130001-135000','135001-140000','140001-145000'])

Price_Group_Distribution_countplot.bar_label(Price_Group_Distribution_countplot.containers[0])

plt.xticks(rotation=90) 
```
